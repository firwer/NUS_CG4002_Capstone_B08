```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
file_path = 'red_debug_leo_my_data.csv'  # Update the path if needed
data = pd.read_csv(file_path)

# Loop over each row (sample) in the dataset
for index, row in data.iterrows():
    # Convert the sensor data from string to lists
    ax = np.array(eval(row['ax']))
    ay = np.array(eval(row['ay']))
    az = np.array(eval(row['az']))
    gx = np.array(eval(row['gx']))
    gy = np.array(eval(row['gy']))
    gz = np.array(eval(row['gz']))
    
    gesture = row['gesture']
    
    # Plot acceleration data (ax, ay, az)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(ax, label='ax')
    plt.plot(ay, label='ay')
    plt.plot(az, label='az')
    plt.title(f'Acceleration (ax, ay, az) for Gesture: {gesture}')
    plt.xlabel('Time Step')
    plt.ylabel('Acceleration Value')
    plt.legend()
    
    # Plot gyroscope data (gx, gy, gz)
    plt.subplot(1, 2, 2)
    plt.plot(gx, label='gx')
    plt.plot(gy, label='gy')
    plt.plot(gz, label='gz')
    plt.title(f'Gyroscope (gx, gy, gz) for Gesture: {gesture}')
    plt.xlabel('Time Step')
    plt.ylabel('Gyroscope Value')
    plt.legend()
    
    # Display the plots
    plt.tight_layout()
    plt.show()
    
    # Optional: Stop after a few samples to avoid excessive output
    # if index >= 4:  # Uncomment to show only first 5 samples
    #    break
```
